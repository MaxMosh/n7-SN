```python
from manim import *

class CurveTransformation(Scene):
    def construct(self):
        # Définir les axes
        axes = Axes(
            x_range=[-3, 3, 1],
            y_range=[-1.5, 1.5, 1],
            axis_config={"include_numbers": True}
        )
        
        points1 = np.random.normal(size=(100, 2))  # 100 points dans R^2
        points2 = points1 * np.array([1.5, 0.5])   # Transformation par exemple
        # Définir les courbes
        curve1 = self.create_curve(points1)
        self.play(Create(curve1))

        # Création de la courbe transformée
        curve2 = self.create_curve(points2)

        # Ajouter les courbes et les axes à la scène
        self.play(Transform(curve1, curve2))
        self.wait(2)

    def create_curve(self, points):
        """ Créer une courbe à partir d'un ensemble de points. """
        curve = VMobject()
        curve.set_points_smoothly([*points, points[0]])  # Relier les points en boucle
        curve.set_color(ORANGE)  # Couleur de la courbe (optionnel)
        curve.set_stroke(width=2)  # Épaisseur du trait (optionnel)
        return curve
# Exécuter le script avec Manim en ligne de commande pour générer l'animation
```

```python
from manim import *
import numpy as np
from scipy.stats import gaussian_kde

class TwoGaussianTransformation(Scene):
    def construct(self):
        # Générer les points
        tirage = []
        for i in range(10000):
            which_curve = np.random.binomial(n=1,p=0.5)
            if which_curve == 0:
                tirage.append(np.random.normal(-3,1))
            else:
                tirage.append(np.random.normal(3,1))
        
        points_dist1 = np.array(tirage)

        #points_dist1 = np.random.normal(loc=0, scale=1, size=1000)

        beta = 0.01
        temp = np.array(tirage)
        point_dist_list = []
        for i in range(1,1000):
            eps = np.random.normal(size=len(tirage))
            temp = np.sqrt(1-beta)*temp + np.sqrt(beta)*eps
            if i%100 == 0:
                point_dist_list.append(temp)
        #points_dist2 = temp

        #points_dist2 = np.random.uniform(low=-3, high=3, size=1000)
        
        # Estimer la densité de probabilité pour les deux distributions
        kde1 = gaussian_kde(points_dist1)
        
        #kde2 = gaussian_kde(points_dist2)
        kde_list = [gaussian_kde(point_dist) for point_dist in point_dist_list]

        # Définir les plages de valeurs pour les courbes
        x_vals = np.linspace(-4, 4, 1000)
        y_vals1 = kde1(x_vals)

        #y_vals2 = kde2(x_vals)
        y_vals_list = [kde_it(x_vals) for kde_it in kde_list]
        
        # Créer les courbes de densité de probabilité
        curve1 = self.get_density_curve(x_vals, y_vals1, color=BLUE)
        
        
        #curve2 = self.get_density_curve(x_vals, y_vals2, color=GREEN)
        curve_list = [self.get_density_curve(x_vals, y_vals_it, color=BLUE) for y_vals_it in y_vals_list]

        # Créer des axes
        axes = Axes(
            x_range=[-4, 4, 1],
            #y_range=[0, max(max(y_vals1), max(y_vals2)) + 0.1, 0.1],
            y_range=[0, 0.5, 0.1],
            axis_config={"color": WHITE}
        ).add_coordinates()

        # Placer la première courbe sur la scène
        curve1.move_to(axes.c2p(0, 0), aligned_edge=DOWN)
        
        for curve_it in curve_list:
            curve_it.move_to(axes.c2p(0, 0), aligned_edge=DOWN)
        #curve2.move_to(axes.c2p(0, 0), aligned_edge=DOWN)

        # Ajouter les axes et la première courbe à la scène
        self.add(axes, curve1)




         # Ajouter une équation fixe
        equation = MathTex(r"\sqrt{1-\beta} \, x + \sqrt{\beta} \, \epsilon").to_corner(UL)
        self.add(equation)

        # Ajouter un compteur d'itérations
        iteration_text = Text("Iteration: 0").to_corner(UR)
        self.add(iteration_text)




        self.wait(1.5)

        # Animer la transformation de la première courbe vers la deuxième
        #self.play(Transform(curve1, curve_list[0]))
        #self.wait(1)
        #for i in range(len(curve_list) - 1):
        #    self.play(Transform(curve_list[i], curve_list[i + 1]))
        #    self.wait(1)
        for i, curve_it in enumerate(curve_list):
            self.play(Transform(curve1, curve_it))
            iteration_text.become(Text(f"Iteration: {(i+1)*100}").to_corner(UR))
            self.wait(0.5)
        self.wait(1)


    def get_density_curve(self, x_vals, y_vals, color):
        # Créer une courbe de densité de probabilité
        points = [np.array([x, y, 0]) for x, y in zip(x_vals, 10.*y_vals)]
        #points = [np.array([x, y, 0]) for x, y in zip(x_vals, y_vals)]
        curve = VMobject()
        curve.set_points_smoothly(points)
        curve.set_color(color)
        return curve
```

```python
from manim import *
import numpy as np
from scipy.stats import gaussian_kde
from sklearn.neighbors import KernelDensity

class TwoUniformTransformation(Scene):
    def construct(self):
        # Générer les points
        tirage = []
        for i in range(10000):
            which_curve = np.random.binomial(n=1,p=0.5)
            if which_curve == 0:
                tirage.append(np.random.uniform(high=-3, low=-1))
            else:
                tirage.append(np.random.uniform(high=1, low=3))
        
        points_dist1 = np.array(tirage)

        #points_dist1 = np.random.normal(loc=0, scale=1, size=1000)

        beta = 0.005
        temp = np.array(tirage)
        point_dist_list = []
        for i in range(1,1000):
            eps = np.random.normal(size=len(tirage))
            temp = np.sqrt(1-beta)*temp + np.sqrt(beta)*eps
            if i%100 == 0:
                point_dist_list.append(temp)
        #points_dist2 = temp

        #points_dist2 = np.random.uniform(low=-3, high=3, size=1000)
        
        # Estimer la densité de probabilité pour les deux distributions
        
        kde1 = KernelDensity(kernel='linear', bandwidth=0.5).fit(points_dist1[:, np.newaxis])
        
        #kde2 = gaussian_kde(points_dist2)
        kde_list = [KernelDensity(kernel='linear', bandwidth=0.5).fit(point_dist[:, np.newaxis]) for point_dist in point_dist_list]

        # Définir les plages de valeurs pour les courbes
        x_vals = np.linspace(-6, 6, 1000)
        y_vals1 = np.exp(kde1.score_samples(x_vals[:, np.newaxis]))
        y_vals_list = [np.exp(kde_it.score_samples(x_vals[:, np.newaxis])) for kde_it in kde_list]
        
        # Créer les courbes de densité de probabilité
        curve1 = self.get_density_curve(x_vals, y_vals1, color=BLUE)
        
        
        #curve2 = self.get_density_curve(x_vals, y_vals2, color=GREEN)
        curve_list = [self.get_density_curve(x_vals, y_vals_it, color=BLUE) for y_vals_it in y_vals_list]

        # Créer des axes
        axes = Axes(
            x_range=[-4, 4, 1],
            #y_range=[0, max(max(y_vals1), max(y_vals2)) + 0.1, 0.1],
            y_range=[0, 0.5, 0.1],
            axis_config={"color": WHITE}
        ).add_coordinates()

        # Placer la première courbe sur la scène
        curve1.move_to(axes.c2p(0, 0), aligned_edge=DOWN)
        
        for curve_it in curve_list:
            curve_it.move_to(axes.c2p(0, 0), aligned_edge=DOWN)
        #curve2.move_to(axes.c2p(0, 0), aligned_edge=DOWN)

        # Ajouter les axes et la première courbe à la scène
        self.add(axes, curve1)



         # Ajouter une équation fixe
        equation = MathTex(r"\sqrt{1-\beta} \, x + \sqrt{\beta} \, \epsilon").to_corner(UL)
        self.add(equation)

        # Ajouter un compteur d'itérations
        iteration_text = Text("Iteration: 0").to_corner(UR)
        self.add(iteration_text)





        self.play(Transform(curve1, curve1))
        self.wait(0.5)

        # Animer la transformation de la première courbe vers la deuxième
        #self.play(Transform(curve1, curve_list[0]))
        #self.wait(1)
        #for i in range(len(curve_list) - 1):
        #    self.play(Transform(curve_list[i], curve_list[i + 1]))
        #    self.wait(1)
        for i, curve_it in enumerate(curve_list):
            self.play(Transform(curve1, curve_it))
            iteration_text.become(Text(f"Iteration: {(i+1)*100}").to_corner(UR))
            self.wait(0.5)
        self.wait(1)


    def get_density_curve(self, x_vals, y_vals, color):
        # Créer une courbe de densité de probabilité
        points = [np.array([x, y, 0]) for x, y in zip(x_vals, 10.*y_vals)]
        #points = [np.array([x, y, 0]) for x, y in zip(x_vals, y_vals)]
        curve = VMobject()
        curve.set_points_smoothly(points)
        curve.set_color(color)
        return curve
```
