```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### **Exercice 2**

```python
file = pd.read_csv("Erreurs.txt", header=None)
```

```python
liste_errL2 = []
liste_errH1 = []
liste_ind = []
for i in range(len(file)):
    temp_string = file.iloc[i,0]
    errL2 = temp_string.split(" ")[5]
    liste_errL2.append(float(errL2))
    errH1 = temp_string.split(" ")[9]
    liste_errH1.append(float(errH1))
    ind = temp_string.split(" ")[1]
    liste_ind.append(int(ind))
```

```python
plt.plot(liste_ind, liste_errL2)
plt.xscale("log")
plt.yscale("log")
```

```python
np.polyfit(np.log(1./(2**np.array(liste_ind))), np.log(liste_errL2), 1)
```

```python
plt.plot(liste_ind, liste_errH1)
plt.xscale("log")
plt.yscale("log")
```

```python
np.polyfit(np.log(1./(2**np.array(liste_ind))), np.log(liste_errH1), 1)
```
